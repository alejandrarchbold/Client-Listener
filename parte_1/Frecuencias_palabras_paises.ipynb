```python
import plotly.express as px
import numpy as np
df = pd.read_excel("freq_total_2.xlsx")
df["Mundo"] = "Mundo" # in order to have a single root node
fig = px.treemap(df, path=['Mundo', 'pais', 'banco'], values='freq',
                  color='freq', #hover_data=['iso_alpha'],
                  color_continuous_scale='RdBu',
                  color_continuous_midpoint=np.average(df['freq'], weights=df['freq']))
fig.show()
```
